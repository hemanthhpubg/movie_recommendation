```python
def convert2(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if(c!=3):
            l.append(i['name'])
            c+=1
        else:
            break
        
    return l
```

```python
import nltk
def tokenizer(i):
    i=i.lower()
    tokens=nltk.word_tokenize(i)
    stop_words=set(stopwords.words('english'))
    tokens=[i for i in tokens if i.lower() not in stop_words]
    special_characters = ['`','~','!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '-', '_', '=', '+', '[', ']', '{', '}', '|', '\\', ';', ':', '\'', '\"', ',', '<', '>', '?']
    tokens=[i for i in tokens if i.lower() not in  special_characters]
    return tokens        
tokenizer("I loved pirated of the caribbean: Recommend movies like it")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline

# Define the dataset of labeled examples
dataset = [
    (r'(best|good|great)? movies (similar|like|based|related) avatar', 'movie recommendation'),
    (r'movies (acted|played) by robert downey', 'cast'),
    (r'movies (directed|made) by james cameron', 'director'),
    (r'(drama|tvmovie|comedy|adventure|western|sciencefiction|science fiction|fantasy|documentary|action|crime|history|family|horror|romance|mystery|thriller|war|foreign|animation|music) based movies', 'genre'),
    (r'(best|good|great)? movies rated', 'rating')
]

# Define the pipeline for the SVM model
model = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('classifier', LinearSVC())
])

# Train the SVM model on the labeled dataset
model.fit([x[0] for x in dataset], [x[1] for x in dataset])


# Use the trained model to classify a new user query
query = 'movies like PIRATES '
intent = model.predict([query])[0]

print('User query:', query)
print('Intent:', intent)
```

```python
import spacy
from spacy.language import Language
from spacy.tokens import Span
import re

# Define the regular expression pattern
pattern1=r'(.)*(loved|liked){1}(\s)*(watching)?(\s)*([a-zA-Z0-9\s]+?)(\s)*(\.|suggest|recommend|best|good|great)+(.)*'
pattern2 = r'(.)*(suggest|recommend)?(\s)*(best|great|good)?(\s)*(movies|films){1}(\s)*(like|similar|related|based){1}(\s)*([a-zA-Z0-9\s]+)(.)*'
pattern3=r'(.*)(movies|films)(.*)(acted|played){1}(\s)*([a-zA-Z0-9\s]+)(.*)'
pattern4=r'([a-zA-Z0-9\s]+)(\s*)(acted|played){1}(.*)(movies|films)(.*)'
pattern5=r'(.*)(movies|films)(.*)(directed|made){1}(\s)*([a-zA-Z0-9\s]+)(.*)'
pattern6=r'([a-zA-Z0-9\s]+)(\s)*(directed|made){1}(.*)(movies|films)(.*)'
pattern7=r'(drama|tvmovie|comedy|adventure|western|sciencefiction|science fiction|fantasy|documentary|action|crime|history|family|horror|romance|mystery|thriller|war|foreign|animation|music)'
pattern8=r'(.)*(movies|films){1}(\s)*(rated above){1}(\s)*([0-9\.]+)'
regex1 =re.compile(pattern1)
regex2=re.compile(pattern2)
regex3=re.compile(pattern3)
regex4=re.compile(pattern4)
regex5=re.compile(pattern5)
regex6=re.compile(pattern6)
regex7=re.compile(pattern7)
regex8=re.compile(pattern8)
def remove_labels(doc):
    doc.ents = [] 
def labelling(doc,match,group,label):
        remove_labels(doc)
        match_text = match.group(group).strip()
        start = doc.text.find(match_text)
        end = start + len(match_text)
        span = doc.char_span(start, end)
        start=span.start
        end=span.end
        print(start,end)
        if start is not None and end is not None and not any(ent.start <= start and ent.end >= end for ent in doc.ents):
            span = Span(doc, start=start, end=end, label=label)
            doc.ents = list(doc.ents) + [span]  
# Create a custom component for the NER model
@Language.component("movie_component")
def movie_component(doc):
    # Apply the regular expression pattern to the document
    match1=regex1.search(doc.text)
    match2=regex2.search(doc.text)
    match3=regex3.search(doc.text)
    match4=regex4.search(doc.text)
    match5=regex5.search(doc.text)
    match6=regex6.search(doc.text)
    match7=regex7.search(doc.text)
    match8=regex8.search(doc.text)
    if match1:
        labelling(doc,match1,6,"MOVIE")
    elif match2:
        labelling(doc,match2,10,"MOVIE")
    elif match3:
        labelling(doc,match3,6,"CAST")
    elif match4:
        labelling(doc,match4,1,"CAST")
    elif match5:
        labelling(doc,match5,6,"DIRECTOR")
    elif match6:
        labelling(doc,match6,1,"DIRECTOR")
    elif match7:
        matches = re.findall(pattern7,doc.text)
        for i in matches:
            start = doc.text.find(i)
            end = start + len(i)
            span = doc.char_span(start, end)
            start=span.start
            end=span.end
            if start is not None and end is not None and not any(ent.start <= start and ent.end >= end for ent in doc.ents):
                span = Span(doc, start=start, end=end, label="GENRE")
                doc.ents = list(doc.ents) + [span]
    return doc
        
# Load a pre-trained English language model and add the custom component
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('movie_component')

# Test the string against the pattern
```

```python
from fuzzywuzzy import fuzz
# Step 2-6: Find the closest match
def find_match(text,arr):
    text = text.lower().strip().replace('.', '').replace(',', '').replace(" ","")  # Normalize input
    best_match = None
    best_score = 0
    c=-1
    for i in arr:
        c+=1
        d=-1
        for j in i:
            d+=1
            text_normalized = j.lower().strip().replace('.', '').replace(',', '')  # Normalize movie name
            score = fuzz.token_sort_ratio(text, text_normalized)  # Compare similarity using token sort ratio
            if score > best_score:
                best_match = j
                best_score = score
                loc1=c
                loc2=d
    print(best_score)
    if best_score>=70:
        return [best_match,best_score,loc1,loc2]
# Example usage:
find_match("harry potter",ml)
    
```
